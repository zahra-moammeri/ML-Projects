```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('datasets/carprices.csv')
df
```

```python
df= df.rename(columns={'Car Model': 'car_model','Sell Price($)':'SellPrice','Age(yrs)':'Age'})
df
```

```python
dummies = pd.get_dummies(df.car_model)
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['car_model', 'Audi A5'], axis='columns')
final
```

```python
reg_model = LinearRegression()
```

```python
X = final.drop('SellPrice', axis='columns')
X
```

```python
y = final.SellPrice
y
```

```python
reg_model.fit(X,y)
```

```python
reg_model.predict([[69000, 6, 1, 0]])
```

```python
reg_model.predict([[45000, 4, 0, 1]])
```

```python
reg_model.predict([[86000, 7, 1, 0]])
```

```python
reg_model.score(X, y)
```

# One Hot Encoding Using SKlearn

```python
df
```

```python
lb = LabelEncoder()
lbdf = df
lbdf.car_model = lb.fit_transform(lbdf.car_model)
lbdf
```

```python
x = lbdf[['car_model', 'Mileage', 'Age']].values
x
```

```python
y = lbdf.SellPrice
y
```

```python
ct = ColumnTransformer([('car_model', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
x[:, 1:]
```

```python
model= LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.predict([[0, 0, 1, 45000, 4]])
```

```python
model.predict([[0, 1, 0, 86000, 7]])
```

```python
model.score(x, y)
```
