```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score


from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
```

```python
data = pd.read_csv('datasets/Student_Performance.csv')
data.head()
```

```python
data.info()
```

```python
data.columns
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.duplicated().sum()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data = data.rename(columns={'Hours Studied': 'study_hours', 'Previous Scores': 'scores',
                     'Extracurricular Activities': 'extra_activity', 'Sleep Hours': 'sleep_hours',
                     'Sample Question Papers Practiced': 'papers_practice', 'Performance Index': 'performance'
                       })
```

```python
data.head()
```

```python
data.describe()
```

```python
label = LabelEncoder()
df = data
df.extra_activity = label.fit_transform(df.extra_activity)
df.head()
```

```python
df.describe()
```

```python
x = data.drop(columns='performance')
x.columns
```

```python
x= x.values
```

```python
y = data['performance']
y
```

```python
col_transform = ColumnTransformer([('extra_activity', OneHotEncoder(), [0])], remainder='passthrough')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
print('x_train shape', x_train.shape); print('x_test shape',x_test.shape)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
r2_score(y_pred, y_test)
```
