```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
%matplotlib inline
```

```python
digits = load_digits()
dir(digits)
```

```python
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
df.head()
```

```python
 X_train, X_test, y_train, y_test = train_test_split(df.drop(['target'], axis = 'columns'), digits.target, test_size = 0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 60)

# n_estimators is the numbers of trees that has been used. the default is 100 
```

ensemble is a term used when using multiple algorithms to predict the outcome

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
sns.heatmap(cm, annot=True)
plt.xlabel('predicted')
plt.ylabel('truth');
```

# On Iris Dataset

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
dir(iris)
```

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
iris_df.head()
```

```python
iris_df['target'] = iris.target
iris_df.head()
```

```python
iris_df['Flower Name'] = iris_df.target.apply(lambda x: iris.target_names[x] )
```

```python
iris_df
```

```python
X = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
iris_model = RandomForestClassifier()
```

```python
iris_model.fit(X_train, y_train)
```

```python
iris_model.score(X_test, y_test)
```

```python
y_predict = iris_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
import seaborn as sns
sns.heatmap(cm , annot=True)
```
