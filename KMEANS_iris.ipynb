```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
iris = load_iris()
iris.keys()
```

```python
iris.feature_names
```

```python
iris.target
```

```python
iris.target_names
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df.head()
```

```python
iris_df['target'] = iris.target
```

```python
iris_df.head()
```

```python
df = iris_df.drop(iris_df[['sepal length (cm)', 'sepal width (cm)']], axis=1)
df.head()
```

```python
plt.scatter(df[['petal length (cm)']], df.target)
plt.scatter(df[['petal width (cm)']], df.target)
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['petal length (cm)', 'petal width (cm)', 'target']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]
df2 = df[df['cluster'] == 2]
```

```python
plt.scatter(df0[['petal length (cm)']], df0.target, color= 'green', label='cluster_0')
plt.scatter(df0[['petal width (cm)']], df0.target, color='green', label='cluster_0')
plt.scatter(df1[['petal length (cm)']], df1.target, color='blue', label='cluster_1')
plt.scatter(df1[['petal width (cm)']], df1.target, color='blue', label='cluster_1')
plt.scatter(df2[['petal length (cm)']], df2.target, color='yellow', label='cluster_2')
plt.scatter(df2[['petal width (cm)']], df2.target, color='yellow', label='cluster_2')
plt.xlabel('Petal')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['petal length (cm)']])
```

```python
df['petal length (cm)'] = scaler.transform(df[['petal length (cm)']])
df.head()
```

```python
scaler.fit(df[['petal width (cm)']])
```

```python
df['petal width (cm)'] = scaler.transform(df[['petal width (cm)']])
df.head()
```

```python
kms = KMeans(n_clusters=3)
y_predict = kms.fit_predict(df[['petal length (cm)', 'petal width (cm)', 'target']])
y_predict
```

```python
df['cluster'] = y_predict
```

```python
df.head()
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]
df2 = df[df['cluster'] == 2]



plt.scatter(df0[['petal length (cm)']], df0.target, color= 'green', label='cluster_0')
plt.scatter(df0[['petal width (cm)']], df0.target, color='green', label='cluster_0')
plt.scatter(df1[['petal length (cm)']], df1.target, color='blue', label='cluster_1')
plt.scatter(df1[['petal width (cm)']], df1.target, color='blue', label='cluster_1')
plt.scatter(df2[['petal length (cm)']], df2.target, color='yellow', label='cluster_2')
plt.scatter(df2[['petal width (cm)']], df2.target, color='yellow', label='cluster_2')
plt.xlabel('Petal')
plt.ylabel('Target')
plt.legend()
plt.show()
```

# Centroids

```python
kms.cluster_centers_
```

```python
df0 = df[df['cluster'] == 0]
df1 = df[df['cluster'] == 1]


plt.scatter(df0[['petal length (cm)']], df0.target, color= 'green', label='cluster_0')
plt.scatter(df0[['petal width (cm)']], df0.target, color='green', label='cluster_0')
plt.scatter(df1[['petal length (cm)']], df1.target, color='blue', label='cluster_1')
plt.scatter(df1[['petal width (cm)']], df1.target, color='blue', label='cluster_1')
plt.scatter(df2[['petal length (cm)']], df2.target, color='yellow', label='cluster_2')
plt.scatter(df2[['petal width (cm)']], df2.target, color='yellow', label='cluster_2')
plt.scatter(kms.cluster_centers_[:,0], kms.cluster_centers_[:,1], color='red', marker='*', label='centroid')
plt.xlabel('Petal')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
kms_rng = range(1, 10)
SSE = []
for rng in kms_rng:
    km = KMeans(n_clusters=rng)
    km.fit(df[['petal length (cm)', 'target']])
    km.fit(df[['petal width (cm)', 'target']])
    SSE.append(km.inertia_)

```

```python
SSE
```

```python
plt.plot(kms_rng, SSE);
```
